```python
import pandas as pd
import numpy as np

oyo_students = pd.read_csv("C:\\Users\\HP\\Downloads\\oyo_state_students.csv")
oyo_students.head(100)
```

```python
highest_students_local_government = oyo_students[oyo_students['Total Students'] == oyo_students['Total Students'].max()]
highest_students_local_government
# This will give us the local government with the highest students
```

```python
lowest_students_local_government = oyo_students[oyo_students['Total Students'] == oyo_students['Total Students'].min()]
lowest_students_local_government
# This will give us the local government with the lowest students
```

```python
total_male_students = oyo_students['PRE-PRY MALE'].sum() + oyo_students['PRY I MALE'].sum() + oyo_students['PRY II MALE'].sum() + oyo_students['PRY III MALE'].sum()
total_male_students
# This will give us the total male students we have in the dataset
```

```python
total_female_students = oyo_students['PRE-PRYFEMALE'].sum() + oyo_students['PRY I FEMALE'].sum() + oyo_students['PRY II FEMALE'].sum() + oyo_students['PRY III FEMALE'].sum()
total_female_students
# This will give us the total female students we have in the dataset
```

```python
import matplotlib.pyplot as plt

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(oyo_students['LOCAL GOVERNMENT'], oyo_students['Student count cartegory'], label="Total Students Group")

# Adding labels and title
ax.set_xlabel('Local Government Area')
ax.set_ylabel('Student Count')
ax.set_title('Student Count by Category across Local Government Areas')
ax.legend()

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

# Show plot
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt


# Define the boundaries of your map
lower_left_lon = 2.5
lower_left_lat = 6.0
upper_right_lon = 4.0
upper_right_lat = 8.0

# Create a map using Basemap
m = Basemap(llcrnrlon=lower_left_lon, llcrnrlat=lower_left_lat,
            urcrnrlon=upper_right_lon, urcrnrlat=upper_right_lat,
            resolution='h', projection='merc')

# Draw coastlines, countries, and states
m.drawcoastlines()
m.drawcountries()
m.drawstates()

# Plotting student counts by category
x, y = m(oyo_students['LOCAL GOVERNMENT'], [0, 1, 2, 3])  # Convert LGAs to map coordinates
sizes = oyo_students['Student count cartegory']
m.scatter(x, y, s=sizes, alpha=0.5, label="Total Students Group")

# Adding labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Student Count by Category across Local Government Areas')
plt.legend()

# Show plot
plt.show()
```
