```python
import requests
from bs4 import BeautifulSoup as soup
website = "https://www.etsy.com/search/home-and-living?is_merch_library=true&mosv=sese&moci=1175624645506&mosi=1198011278141&q=home+gift&ref=marketing_pages_show_more&page=1"
response = requests.get(website)
response
```

```python
page_soup = soup(response.text, "html.parser")

page_soup
```

```python
items = page_soup.findAll("div", {"class": "v2-listing-card__info"})
items
```

```python
for i in items:
    itmname = i.findAll("h3", {"class":"wt-text-caption v2-listing-card__title wt-text-truncate"})
    itmprice = i.findAll("span", {"class":"wt-screen-reader-only"})
    print(itmname[0].text.strip())
    print(itmprice[2].text.strip()[12:])
```

```python
itemnames = []
itemprices = []

for i in items:
    itmname = i.findAll("h3", {"class":"wt-text-caption v2-listing-card__title wt-text-truncate"})
    itmprice = i.findAll("span", {"class":"wt-screen-reader-only"})
    author = i.findAll("div", {"class":"wt-display-flex-xs"})
    print(author[0].findAll("span", {"aria-hidden":"true"})[-1].text)
#     itemnames.append(itmname[0].text.strip())
#     itemprices.append(itmprice[2].text.strip()[12:])
    if len(author) == 0:
        pass
    else:
        print(author)
```

```python
import pandas as pd
etsyProducts = pd.DataFrame({"Product Name":itemnames, "Product Price":itemprices})
etsyProducts.head()
etsyProducts.to_csv("etsyproducts.csv")
```
