```python
import requests
from bs4 import BeautifulSoup as soup
website = "https://www.etsy.com/c/gifts/mothers-day-gifts?ref=MDAY24_cat_nav"
response = requests.get(website)
response.text
```

```python
page_soup = soup(response.text, "html.parser") #is an argument passed to the soup function, specifying the parser to be used to parse the HTML. It uses the built in HTML parses provided by BeautifulSoup

page_soup
```

```python
items = page_soup.findAll("div", {"class": "b-list-advert__gallery__item js-advert-list-item"})
# itprice = items[0].findAll("div", {"class": "qa-advert-price"})
# itprice[0].text.strip()
items
```

```python
import pandas as pd
itemnames = []
itemprices = []
for obj in items:
    itp = obj.findAll("div", {"class": "qa-advert-price"})
    itn = obj.findAll("div", {"class": "b-advert-title-inner qa-advert-title b-advert-title-inner--div"})
    itemprices.append(itp[0].text.strip())
    itemnames.append(itn[0].text.strip())

konga_products = pd.DataFrame({"Name": itemnames, "Price": itemprices})
konga_products
```

```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
web = "https://en.wikipedia.org/wiki/States_of_Nigeria"
resp = requests.get(web)
resp
```

```python

page_soup = soup(resp.text, "html.parser") 

page_soup
```

```python
page_cont = page_soup.findAll("div", {"class":"mw-page-container"})
page_cont
```

```python
table = page_cont[0].findAll("table", {"class": "wikitable sortable plainrowheaders"})[0]
table
```

```python
image_sources = []
for row in table.find_all('tr'):
    img_tag = row.find('img')
    if img_tag:
        img_src = img_tag.get('src')
        image_sources.append(img_src)

# Print the extracted image sources
for src in image_sources:
    print(src)
```

```python
names = []
for row in table.find_all('tr'):
    th_tag = row.findAll("th", {"scope":"row"})
    if th_tag:
        names.append(th_tag[0].text.strip())
```

```python
import pandas as pd

df = pd.DataFrame({"Name":names, "Image":image_sources})
df.to_csv("Imageswithstate")
```
