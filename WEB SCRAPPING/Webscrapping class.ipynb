Web scraping is the process of automatically extracting data from websites. It involves using software or programming code to access web pages, retrieve the desired information, and store it for further analysis or use. Web scraping can be used for various purposes, such as data collection, market research, price comparison and more. However, it's essential to respect website terms of service and legal regulations when engaging in web scraping activities.

Web scraping using Beautiful soup
Web scraping means to collect datat from the internet. As a beginner in data science, you must have seen CSV files on the internet distributed by some popular website e.g kaggle and other govt websites. The data is prepared by either collecting, writing, using standard methods or by scrapping it from the internet web scrapping to create a CSV file so we need two primary packages for this task, BeautifulSOup, urllib and selenium. We can easily install both of these packages using the pip command -pip install bs4, pip install urllib, pip install selenium.

Beautiful Soup is a Pythpon library used for web scraping purposes. It simplifies the process of parsing HTMl and XML documents, making it easier to extract data from web pages. Beautifl Soup creates a parse tree from the HTML source code of a web page, which you can then navigate and search to extract specific elements or information. Beautiful Soup relies on the structure of the HTML or XML documents on a website. It works best when the website's HTML is well-formed and follows standard conventions. Websites with complex or non-standard structures may require more advanced techniques or additional libraries.

However, Websites that heavily rely on JavaScript to load content dynamically may pose challenges for Beautiful Soup. Since Beautiful Soup parses static HTML content, it may not easily access data loaded via Javascript. In such cases, you might need to combine Beautiful Soup with a tool like Selenium for web scrapping.

```python
# from bs4 import BeautifulSoup as soup
# from urllib.request import urlopen as uReg #urlopen is used for opening and reading URLs
# # my_url = "https://www.flipkart.com/samsung-mobile-store?otracker=nmenu_sub_Electronics_0_Samsung"
# my_url = "https://www.my9jarocks.info/category/videodownload"
# uClient = uReg(my_url)
# page_html = uClient.read() #reads the content of the webpage
# uClient.close()

import requests
from bs4 import BeautifulSoup as soup
website = "https://gutenberg.org/ebooks/bookshelf/151/"
response = requests.get(website)
response
```

```python
page_soup = soup(response.text, "html.parser") #is an argument passed to the soup function, specifying the parser to be used to parse the HTML. It uses the built in HTML parses provided by BeautifulSoup

page_soup
```

```python
containers = page_soup.findAll("div", {"id": "container"})
print(len(containers))
print(soup.prettify(containers[0]))
```

```python
contents =  containers[0].findAll("div", {"class": "_3YgSsQ"})
len(contents)
```

```python
container = contents[0]
product_name = container.findAll("a", {"class": "s1Q9rs"})
product_name[0].text
```

```python
price = container.findAll("div", {"class": "_30jeq3"})
print(price[0].text)
```

```python
ratings = container.findAll("span", {"class": "_2_R_DZ"})
print(ratings[0].text)
```

```python
import pandas as pd
import numpy as np

products = []
pricing = []
rate = []
for container in contents:
    product_name = container.findAll("a", {"class": "s1Q9rs"})
    name = product_name[0].text
    products.append(name)
    price_container = container.findAll("div", {"class": "_30jeq3"})
    if len(price_container) == 0:
        pricing.append(np.nan)
    else:
        price = price_container[0].text
        pricing.append(price)
    rating_container = container.findAll("span", {"class": "_2_R_DZ"})
    if len(rating_container) == 0:
        rate.append(np.nan)
    else:
        rating = rating_container[0].text
        rate.append(rating)
    
    print("Product_Name:" + name)
    print("Price:" + price)
    print("Ratings:" + rating)
```

```python
df = pd.DataFrame({"Product Name":products, "Price":pricing, "Rating":rate})
df.to_csv("Samsung_mobile.csv")
samsung = pd.read_csv("Samsung_mobile.csv")
samsung
```

```python
import urllib
my_url = "https://www.flipkart.com/television-store?otracker=nmenu_sub_TVs%20and%20Appliances_0_Televisions&otracker=nmenu_sub_TVs%20%26%20Appliances_0_Television"
header = {'User-Agent':
          'Chrome/23.0.1271.64 Safari/537.11',
          'Accept': 'text/html,application/xhtml,application/xml;q=0.9,*/*;q=0.8', #It indicates that the script can accept various formats of HTML content.
          'Accept.Charset': '150-8859-1,utf-8,q=0.7,*;q=0.3', #handles ISO-8859-1 and utf-8 character encoding.
          'Accept-Encoding': 'none',
          'Accept-Language': 'en-US, en;q=0.8', #script doesn't accept compressed content
          'Connection': 'Keep-alive'} # to maintain a persistent connection
uClient = urllib.request.Request(my_url, headers = header)
page = uReg(uClient).read()
```

```python
# user-Agent: This header simulates the user agent of a web browser. It tells the web server what kind of browser is making the request
# Accept: This header specifies the types of content that the client (script) can understand
# Accept-Charset: This header specifies the character sets that your script can handle. It indicates that your scipt can handle ISO-8859-1 and utf-8 character encodings.
# Accept-Encoding: This header indicates whether your script can handle content compression
# Accept-Language: This header specifies the preferred language of the client
# Connection: This header specifies the type of connection to be used to keep the connection alive.    
```

```python
import yfinance as yf

import datetime
from datetime import date, timedelta
```

```python
today = date.today()

end_date = today.strftime("%Y-%m-%d")

# we will create date interval to collect the stock price of data from today to the last 360 days
d2 = date.today() - timedelta(days=360)
start_date = d2.strftime("%Y-%m-%d")
```

```python
data = yf.download("TMO", start=start_date, end=end_date, progress=False)
```

```python
data.head()
```

```python
# In this datatset, Date is not a column. It's the index of this dataset. To use this data for any data science task, we need to
#convert this index into a column. Below is how you can do that
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
data.head()
```
