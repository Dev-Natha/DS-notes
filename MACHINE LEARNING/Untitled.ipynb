```python
import pandas as pd
melb = pd.read_csv("C:\\Users\\HP\\Documents\\DATASCIENCE\\Dataset\\Melbourne_housing_FULL.csv")
melb
```

```python
melb.info()
```

```python
melb.Price = melb.Price.fillna(melb.Price.mean())
```

```python
melb.corr(method="pearson")
```

```python
feature = melb[["Rooms", "Bedroom2", "Bathroom"]]
feature.Bedroom2 = feature.Bedroom2.fillna(feature.Bedroom2.mean())
feature.Bathroom = feature.Bathroom.fillna(feature.Bathroom.mean())

target = melb.Price
target
```

```python
print(feature.shape, target.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
answer = lr.predict(X_test)
answer
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

accuracy = r2_score(y_test, answer) 
accuracy
```
