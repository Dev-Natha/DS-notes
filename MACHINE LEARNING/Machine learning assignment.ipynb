```python
import pandas as pd
melb = pd.read_csv("C:\\Users\\HP\\Documents\\DATASCIENCE\\Dataset\\Melbourne_housing_FULL.csv")
mel_price = melb.Price.fillna(melb.Price.mean())
melb_room = pd.DataFrame(melb.Rooms, columns=["Rooms"])
melb_bed = melb.Bedroom2.fillna(melb.Bedroom2.mean())
melb.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(melb_room, melb_bed, test_size = 0.2)
X_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
# y_pred.shape
# y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
intercept = lr.intercept_
print(r2)
print(mse)
print(mae)
print(intercept)
```

```python
# How the landsize, buildingarea, newsuburb and distance affect the price

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
melb['newsuburb'] = le.fit_transform(melb["Suburb"])
melb.BuildingArea = melb.BuildingArea.fillna(melb.BuildingArea.mean())
melb.Landsize = melb.Landsize.fillna(melb.Landsize.mean())
melb.Distance = melb.Distance.fillna(melb.Distance.mean())

allfeature = melb[["Landsize", "BuildingArea", "newsuburb", "Distance"]]
```

```python
import pandas as pd
titanic = pd.read_csv("C:\\Users\\HP\\Documents\\DATASCIENCE\\Dataset\\Titanic-Dataset.csv")
titanic.drop(["Name", "Ticket", "Cabin"], axis=1, inplace=True)
titanic["Embarked"] = titanic["Embarked"].fillna(method="ffill")
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].mean())
titanic.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for val in titanic.columns:
    if titanic[val].dtype == object:
        titanic["New " + val] = le.fit_transform(titanic[val])
        titanic.drop(val, axis = 1, inplace=True)
titanic
```

```python
from sklearn.decomposition import PCA 

model = PCA(n_components=3)
model.fit(titanic.drop("Survived", axis=1))
X_2D = model.transform(titanic.drop("Survived", axis=1))

from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_2D, titanic["Survived"], random_state=1, train_size=0.8)

Xtrain.shape

from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB() # 2. instantiate model
model.fit(Xtrain, ytrain) # 3. fit model to data
y_model = model.predict(Xtest) # 4. predict on new data

# print("Predicted \n", y_model)
# print("Original label \n", ytest.values)
Xtest
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```
